# Run drilling simulation

```python
from sub_compute_kick_behavior import compute_kick_behavior

bhp = 8750 # bottom hol pressure when experiencing kick (psi)
kick_molar_mass =  16.04  # molor mass (g/mol)
r_g = kick_molar_mass / 28.96  # specific gravity (fraction)
mud_density = 14.5 # PPG
kill_rate =  180 # gpm
formation_temperature_grad =  1.2 # F/100f 
result =  compute_kick_behavior(bhp, 
                                kick_molar_mass, 
                                mud_density, 
                                kill_rate, 
                                formation_temperature_grad, 
                                r_g = r_g).drop('index', axis = 1)
```

# visualize the output

```python
result.columns
```

```python
import matplotlib.pyplot as plt
result.plot(kind = 'line',
            x = 'time',
            y = 'kick_volume')
plt.minorticks_on()
plt.grid(which = 'major', linestyle = '-', linewidth = 1)
plt.grid(which = 'minor', linestyle = ':', linewidth = 0.5)
plt.ylabel('kick volume, bbls')
plt.xlabel('time, sec')
plt.title('Kick volume vs. time')
```

```python
result.plot(kind = 'line',
            x = 'time',
            y = 'choke_pressure')
plt.minorticks_on()
plt.grid(which = 'major', linestyle = '-', linewidth = 1)
plt.grid(which = 'minor', linestyle = ':', linewidth = 0.5)
plt.ylabel('chock pressure, psi')
plt.xlabel('time, sec')
plt.title('Chock pressure vs. time')
```
